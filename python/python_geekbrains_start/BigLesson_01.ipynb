### зачем нужны условные операторы?
давайте вспомним сказку которая звучит:
- начало сказки
- направо пойдёшь - конч потеряешь
- налево пойдёшь - жизнь потеряешь
- прямо пойдёшь - дальше уйдёшь

есть начало а затем есть развилка. затем какие то действия и сказка заканчивается
в программе тоже самое: в зависимости от какого то условия у нас могут быть разные варианты действий.
чтобы реализовать данную логику в программе понадобится оператор **if**
```
 - начало сказки
 - if налево пойдёшь:
          коня потеряешь
 - конец сказки
```

````python
age10 = int(input('А теперь возраст попугайчика кеши?'))
````

если возраст меньше 18 лет вызвать на экран "доступ запрещен"

````python
if age10 < 18:
    print('В доступе отказанно! только для совершеннолетних!')
else:
    print('ДОСТУП! открыто')
````

у нас есть разветвление **if - else**
**if** проверять будем логическое условие age10 < 18 двоеточие и на новой строчке С ОТСТУПОМ В 4 ПРОБЕЛА
используем отступ в 4 пробела. и пишем условие которое выполняется в случае ВЫПОЛНЕНИЯ условия. мы попадём во внутренний код

````python
print('Какие то действия')
print('end')
````

если условия не выполняются мы не попадаем в блок кода внутри условия print('В доступе отказанно! только для совершеннолетних!')

### Блоки и отступы в коде. 
блок кода - логически сгруппированный набор команд. В пайтоне нет знаков разделяющих блоки, специальных разделителей.
Поэтому обязательно делаем отступ для каждого блока кода.
Отступ 4 пробела ()клавиша tab в paycharm)
```
if - elif - else
if условие:
   код 1
elif условие:
   код 2
else условие:
    код 3
```
Условий **elif** может быть сколько угодно

````python
if age10 < 18:
    print('В доступе отказанно! только для совершеннолетних!')
elif age10 == 18:
    print('Вам ровно 18')
    print('Что мне с вами делать?')
elif age10 > 18 and age10 < 25:
    print('ВЫ попадаете в категорию молодёжь')
else:
    print('ДОСТУП! открыто')
````

Внутрь оператора **if** может быть вложен другой оператор **if**
таких вложений может быть сколько угодно много

````python
if age10 < 18:
    print('В доступе отказанно! только для совершеннолетних!')
elif age10 == 18:
    print('Вам ровно 18')
    print('Что мне с вами делать?')
elif age10 > 18 and age10 < 25:
    print('ВЫ попадаете в категорию молодёжь')
else:
    print('ДОСТУП! открыто')
    if age10 % 5 == 0:
        print('Поздравляем, у вас в этом году юбилей')
````

рассмотрим реализацию с циклом **while** - пока условие выполняется (TRUE): делать определённый набор действий
```
while условие:
   действие 1
   действие 2
```

````python
name5 = input('КТО создатель пайтона ?')

while name5 != 'Гвидо':
    print('НЕ ВЕРНО. лезь и гугли!')
    name5 = input('КТО создатель PYTHON?')

print('Искать умеете! Гвидо! Правильный ответ знаете!')
````

Пользователь будет вводить нам кто создатель пайтон. После этого мы воспользуемся конструкцией **while**
после этого нужно написать условие. пока условие истино мы будем заходить в цикл,
пока name5 не равняется Гвидо т.е если пользователь вводит неверный ответ мы пишем сообщение - не верно лезь и гугли.
и после этого снова просить ввести его кто создатель пайтона.
и когда выходим из цикла и угадали правильный ответ - выводим сообщение о верном ответе

 - вывести числа от 0 до n, где n вводит пользователь

````python
number = 0
n = int(input('Введите любое целое число'))

while number <= n:
    print(number)
    number += 1
````

 - вывод чётных чисел от 0 до n

````python
number = 0
n = int(input('Введите любое целое положительное число'))

while number <= n:
    if number % 2 == 0: # если число в переменной нумбер2 чётное - т.е делится на 2 без остатка
        print(number) # условие чётности выполняется - выводим на экран, не выполняется - прибавляем единичку
    number += 1
````

условие не выполняется , плюсуем единичку и снова отправляем в цикл на проверку.
для нечётных чисел

````python
number = 0
n = int(input('Введите любое число'))

while number <= n:
    if number % 2 != 0:
        print(number)
    number += 1
````

### следующая управляющая конструкция - continue
- переход на новую итерацию цикла(команды в цикле после continue не выполняются)

````python
number = 0
n = int(input('Введите n '))

while number <= n:
    if number % 2 == 0:
        number += 1
        continue
    print(number)
    number += 1
````

Если у нас число чётное мы можем увеличить его на единичку и после сделать continue
а оперцию print(number1) выносим уже после команды континиум.
Если у нас число чётное, тогда мы с вами прибавляем к нему единичку, после этого делаем continium таким образом мы не будем выполнять функцию принт не будем печатать чётное число
соответственно если число нечётное тогда мы делаем принт и нужно обязательно ещё раз добавить в конце единичку
с помощью оператора continium чётные числа мы пропускаем. т. е не будем делать принт и выполнять команду, а сразу переходим на следующую иттерацию цикла и снова идём в тело таким образом мы чередуем четное - принт и нечетное continue.

###  while - else
в блоке else после цикла while мы выполняем действия после того как вышли из цикла while когда условие цикла стало неверным False

````python
number2 = 0

while number2 <= 100:
    print(number2)
    number2 += 1
else:
    print('end')
````

мы просто выводим числа на экран терминала. Запускаем пример получаем числа от 0 до 100.
если допишем в конец файла блок else, обязательно далее делаем отступ 4 пробела, то мы можем писать код который будет выполняться после того как мы вышли из цикла и условие стало ложным.
блок else выполнится только тогда, когда мы выйдем из цикла по условию
это означает, что если мы воспользуемся конструкцией break и заранее выйдим из цикла, то блок else выполняться у нас не будет.

````python
while number2 <= 100:
    print(number2)
    number2 += 1
    if number2 == 33:
        break
else:
    print('else - end')

print('end')
print("!!!ПОЗДРАВЛЯЮ ВЫ ЗАКОНЧИЛИ BIGLESSON 1!!!")
````

выполнился только последний print('end')
т. к мы вышли не по условию цикла number2 <= 100:
блок else не будет выполняться
