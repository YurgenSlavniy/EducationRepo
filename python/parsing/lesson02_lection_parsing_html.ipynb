### атрибут id
Суть атрибута - выделить элемент на странице и сделать его уникальным. Представим мы приходим на вебсайт и делаем на него GET запрос. Нам возвращается данная HTML страничка. Допустим нужная нам информация находится внутри данного контейнера div
```
<div id='d'>
    <p class='red paragraph'>Первый параграф</p>
    <p id="clicable" class='paragraph'>
        <a href='http://www.mail.ru'>Второй <b>кликабельный</b> параграф</a>
    </p>
    <p class='red paragraph left'>Третий параграф</p>
</div>
```

и у этого контейнера div есть атрибут id='d'. Мы можем выбрать данный тег div с использованием выборки по атрибуту id со значением id='d'. Дело в том что другого такого div id='d' не существует и не может существовать ни при каких обстаятельствах. Мы можем быть однозначно быть уверенны что это единственный элемент на странице. 

Обычно в id заключают какие то критически важные вещи. Например поле для ввода логина и пароля. Отдельный айдишник для логина, отдельный для пароля. Выделяют какие то особенные элементы на странице. 

Хоть с технической точки зрения сайт с 2умя одинаковыми айдишниками может открываться и работать, Но если краулер войдёт на наш сайт и это увидит, можно попасть в чёрный список на всю жизнь.

Мы все видили как выглядит код страницы. Её html код это большая такая портянка. По факту для пайтон это воспринимается как одна большая строка. И в этой строке нам надо найти какие то данные. 
Вебразработчики не работают напрямую с html кодом в виде строки. При сильной заморочке можно придумать регулярные выражения которые будут искать нужные элементы достаточно эффективно. Но в целом это очень неудобно. 

Были придуманы более удобные спсобы для обработки html. Это называется DOM. суть заключается в том, что каждый тег обворачивается в некий объект. То есть object Model - Объектная модель. И между объектами строятся связи в зависимости от вёрстки html. для любого сайта, для любой страницы принцип вёрстки одинаков. Берётся изначально первый контейнер - первый встречный тег.

```
<body>
    <div id='d'>
        <p class='red paragraph'>Первый параграф</p>
        <p id="clicable" class='paragraph'>
            <a href='http://www.mail.ru'>Второй <b>кликабельный</b> параграф</a>
        </p>
        <p class='red paragraph left'>Третий параграф</p>
    </div>
    
    <div id='d2'>
        <p><a href="#">Четвёртый параграф</a></p>
        <p>Пятый параграф</p>
    </div>
    
    <p>Шестой параграф</p>
    <p>Седьмой и последний параграф</p>
</body>
```

первый встречный тег у на это `<body>`. Раз тег встретился для него создаётся объект.  Дальше мы идём внутрь и видим, что внутри этого объекта следующий тег `<div id='d'>` и создаётся ещё объект - дочерний объект, который является ребёнком тега `<body>`. Между ними строится связь сверху вниз.  Но внутри тэга  `<div id='d'>` есть ещё теги , которые также объекты для локиги построения DOM. 
`<p class='red paragraph'>`, `<p id="clicable" class='paragraph'>`.
И что самое важное. в DOM всё является объектом. Без исключения. даже комментарий. И коечно текстовое содержимое внутри тегов. Это тоже объект, но просто другого класса.  И он тоже будет учавствовать в полной иерархии, структуре. И родитель у них тоже будет общий. концепция с классическим ООП программированием. 

Даже в пайтон все создаваемые нами классы - наследнике общего класса `object`. Такая концепция в принципе везде. Есть один общий родитель. В нашем примере есть также один общий родитеь - `Document` от которого и растут все ветви. Так общий родитель у этих объектов есть какая то общая модель поведения. Есть определённые методы одинаковые для всех, есть некоторые свойства. 
И соответственно есть различия в зависимости от того с чем они работают. Даже атрибуты тегов тоже являются объектами DOM. 

Существуют специальные парсеры, которые на вход принимают эту самую строку с html кодом, её перерабатывают и на выходе выдают модель. объектную модель - дерево из объектов, которые между собой связаны.
парсер идёт и всё что попадается обварачивает: объект. объект. объект...и все это соединино между собой. 

И когда у нас уже не сырая строка, а объекты конкретных классов, которые пддерживают конкретные методы,  у которых есть конкретные свойства - нам гораздо проще с ними работать. мы можем проитерироваться по объектам, прочитать у них определённое одно свойство, можем повызывать методы у этих объектов и получить результаты. Интерактивность повышается. например у каждого объекта можем прочитать его свойство текст и найти нужный нам текст, который нам по задаче. Или прочитать атрибуты класса. У большей части элементов будет атрибут класс и найти те теги у которых будет нужный нам класс.

### Связи между объектами
Связи очень важная вещь. Когда есть иерархия объектов и мы знаем как она построена,  а мы знаем, так как анализируем html код в инструментах разработчика, и эту иерархию видно. Мы видим в инструментах разработчика вложенные связи. Как в пайтоне отступы. Правее и ниже относится к тому что левее и выше. 
```
    <div id='d'>
        <p class='red paragraph'>Первый параграф</p>
        <p id="clicable" class='paragraph'>
            <a href='http://www.mail.ru'>Второй <b>кликабельный</b> параграф</a>
        </p>
        <p class='red paragraph left'>Третий параграф</p>
    </div>
```

У объекта `<div>` с атрибутом id='d' есть три дочерних объекта `<p>`. Внутренняя вложенность может быть достаточно большой. Когда видим эту взаимосвязь можем оптимизировать работу своих скраперов. 
Одна из часто встречаемых задач.

`<p>Параграф этого <em>текста</em> - это просто пример</p>`

> У нас есть объект P. Это контейнер, который содержит в себе ещё контейнер - тег `<em>текста</em>` . Получается у нас как матрёшка - объект в объекте. Наша цель собрать весь текст. 

Мы при анализе данного html кода вынуждены сначала взять частичку текста "Параграф этого", дальше понимаем что нам надо провалиться вниз в контейнер `<em>текста</em>` и взять оттуда "текста", и снова вернуться изабрать оставшийся текст "- это просто пример". 

Если бы связи между объектами не существовало нам бы пришлось целиком и польностью опросить всю страницу, чтобы найти этот тег `<p>`. И второй раз пришлось бы с самого начала опросить всю страницу, чтобы найти тег `<em>`. Пришлось бы дважды парсить весь документ html.

Намного эффективней воспользоваться связями. Есть методы позволяющие сразу целиком собрать весь текст, избегая внутренних провалов и подъёмов к элементам по этим связям.  Если имеется возможность не вызывать метод для поиска элемента, а воспользоваться родственными связями, то надо пользоваться родственными связями. Это быстрее. 

Когда мы получаем response ответ он хронится в виде текста который где то в памяти лежит. Парсер когда проходит формирует дерево объектов. И всё этодерево также находится в памяти. Постоянно там висит.

### Как устроено дерево в разрезе Beautiful Soup
У тэга `a` есть свойство `parent` мы его раскрываем, там содержится объект который является объектом класса tag. Мы его раскрываем и раз это tag то у него тоже есть `parent`. Это тоже tag и у него соответственно тоже есть `parent`. Мы уже на тэге `<boby>`. Раскрываем и смотрим его родителя `<html>`, но это тоже тег и его родитель `<beautiful_soup>` - дошли до класса от которого всё наследуется. Это уже экземпляр класса beautiful_soup.
