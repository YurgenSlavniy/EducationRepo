## Масштабирование признаков <a class='anchor' id='0'>

В машинном обучении при работе с линейными моделями полезной является практика _масштабирования признаков_. Многие методы машинного обучения, в том числе и линейные, наиболее эффективны в том случае, когда признаки имеют одинаковый масштаб. По сути масштабирование означает приведение признаков к какой-то единой шкале. 

Существует большое количество методов масштабирования, наиболее популярными из которых являются _нормализация_ и _стандартизация_.

Метод **нормализации** заключается в приведении признаков к масштабу в диапазоне [0-1].

Для его реализации необходимо найти минимальное $min_{j} (x^{j}_{i})$ и максимальное $max_{j} (x^{j}_{i})$ значение признака на обучающей выборке. При этом отмасштабированное значение признака будет находиться по формуле

$$x^{j}_{i} = \frac{x^{j}_{i} - min_{j} (x^{j}_{i})}{max_{j} (x^{j}_{i})-min_{j} (x^{j}_{i})}.$$

После преобразования значений признаков минимальное значение превратится в 0, а максимальное - в 1.

**Стандартизация** заключается в получении своего рода значения сдвига каждого признака от среднего. Для ее реализации необходимо вычислить среднее значение признака 

$$\mu_{j} = \frac{1}{l}\sum^{l}_{i=1}x^{j}_{i}$$

и стандартное отклонение, которое находится путем суммирования квадратов отклонения значений признака на объектах выборки от среднего $\mu_{j}$ и делением на число объектов выборки с последующим извлечением корня:

$$\sigma_{j} = \sqrt{\frac{1}{l}\sum^{l}_{i=1}(x^{j}_{i}-\mu_{j})^{2}}$$

Чтобы отмасштабировать признак, каждое его значение преобразуется по формуле

$$x^{j}_{i}=\frac{x^{j}_{i} - \mu_{j}}{\sigma_{j}}.$$

Масштабирование является важным этапом подготовки данных перед применением методов машинного обучения. 

Важным и последним свойством масштабирования является факт, что после масштабирования признаков в линейных моделях веса при них могут интерпретироваться как мера значимости этих признаков.

Существуют различные ситуации, когда целесообразно применять тот или иной метод масштабирования. Нормализовать полезно признаки, опирающиеся на величину значений - такие как расстояние (knn, k-means). Стандартизировать полезно признаки для модели, которая опирается на распределение (линейные модели). В общем случае, когда выбор метода неочевиден, полезной практикой считается создавать масштабированные копии набора данных, с которыми работает специалист, и сравнивать друг с другом полученные после применения модели результаты для выявления оптимального метода масштабирования для имеющейся ситуации.

### Запись лекции 
В данных могут быть очень маленькие признаки и очень большие признаки и соответственно при маленьком признаке веса будут также маленькими, а у признака с большим разбросом знаений (например один признак изменяется от 0 до 1, а второй признак от 0 до 1000) у них разный масштаб и веса также могут быть разного масштаба. В этом случае градиентный спуск будет ходить зигзагом, его будет мотать, будет идти не на прямик к идеальной точке нашего схождения, когда моделька обучена. Для более гладкого обучения мы можем делать масштабирование данных. В этом случае наши признаки и в следствии их веса приходят к одному масштабу. 

Есть основные 2 способа масштабирования данных:
- нормализация. Для каждого признака ищем минимальное и максимальное значение и затем по формуле каждый признак приводим в диапазон от 0 до 1. 
$$x^{j}_{i} = \frac{x^{j}_{i} - min_{j} (x^{j}_{i})}{max_{j} (x^{j}_{i})-min_{j} (x^{j}_{i})}.$$
нормализовав все признаки ои будут изменяться от 0 до 1. и на оси координат будут одинаковые диапазоны по осям от 0 до 1. 

- стандартизация. находится среднее по признаку и по формуле найти стндартное отклонение
$$\sigma_{j} = \sqrt{\frac{1}{l}\sum^{l}_{i=1}(x^{j}_{i}-\mu_{j})^{2}}$$ 
и после поформуле для каждого признака вычисляем масштабируемый признак. 
$$x^{j}_{i}=\frac{x^{j}_{i} - \mu_{j}}{\sigma_{j}}.$$

После выполнения стандартизации среднее будет в 0, а стандартное отклонение в единичке. 

Нормализация лучше себя зарекомендовала в подходах связанных с растояниями (кластеризация или К-ближайших соседей), стандартизация лучше показывает в линейных подходах (линейная модель, линейная регрессия, логистическая регрессия, метод опорных векторов). 

## Стохастический градиентный спуск <a class='anchor' id='1'>

1. Инициализация w

2. Цикл по k = 1,2,3,...:

    * $w^{k} = w^{k-1} - \eta_{k}\nabla Q(w^{k-1}, X)$

    * Если $||w^{k} - w^{k-1}|| < \epsilon$, то завершить.


Вспомним метод градиентного спуска, рассмотренный ранее.

На каждой итерации приближение получается вычитанием из предыдущего вектора градиента, умноженного на некоторый шаг:


$$w^{k} = w^{k-1} - \eta_{k}\nabla Q(w^{k-1}, X).$$

При этом выражение градиента в матричной форме выглядит так:

$$\nabla_{w}Q(w,X) = \frac{2}{l}X^{T}(Xw-y).$$

Если расписать $j$-ю компонетну этого градиента, то получим

$$\frac{\partial Q}{\partial w_{j}} = \frac{2}{l}\sum^{l}_{i=1}x^{j}_{i}(\left \langle w,x_{i} \right \rangle - y_{i}),$$

то есть суммирование по всем $l$ объектам обучающей выборки. Здесь выражение под суммой показывает, как нужно изменить $j$-й вес, чтобы как можно сильнее улучшить качество __на объекте $x_{i}$__, а вся сумма показывает, как нужно изменить вес, чтобы улучшить качество на __всей выборке__.

В этой формуле отражен один из главных недостатков градиентного спуска: если выборка большая по объему, то даже один шаг градиентного спуска будет занимать много вычислительных ресурсов и времени.

Стремление к оптимизации процесса привело к появлению _стохастического градиентного спуска_ (Stochastic gradient descent, SGD). Идея его основана на том, что на одной итерации мы вычитаем не вектор градиента, вычисленный по всей выборке, а вместо этого случайно выбираем один объект из обучающей выборки $x_{i}$ и вычисляем градиент только на этом объекте, то есть градиент только одного слагаемого в функционале ошибки и вычитаем именно этот градиент из текущего приближения вектора весов:

$$w^{k} = w^{k-1} - \eta_{k}\nabla Q(w^{k-1}, \{x_{i}\}),$$

то есть $\nabla Q(w^{k-1}, X)$ заменяется на $\nabla Q(w^{k-1}, \{x_{i}\})$.

Если в случае градиентного спуска мы стараемся на каждой итерации уменьшить ошибку на всей выборке, и по мере увеличения числа итераций ошибка падает монотонно, то в случае стохастического градиентного спуска мы уменьшаем на каждой итерации ошибку только на одном объекте, но при этом есть вероятность увеличить ее на другом объекте, поэтому график изменения ошибки может получаться немонотонным, и даже иметь пики (см. пример по ссылке [1] из списка литературы). То есть на какой-то итерации мы можем даже увеличить ошибку, но при этом в целом по ходу метода ошибка снижается, и рано или поздно мы выходим на нормальный уровень.

1. Инициализация w

2. Цикл по k = 1,2,3,...:

    * Выбираем случайные объект $x_{i}$ из X
    * $w^{k} = w^{k-1} - \eta_{k}\nabla Q(w^{k-1}, \{x_{i}\})$
    * Если $||w^{k} - w^{k-1}|| < \epsilon$, то завершить.

### Запись лекции
Обучаемся на одном объекте за раз. Стахостический градиентный спуск. считаем градиент только по одному объекту. Начало метода такоеже: 

- инициализация весов (либо нулевые, либо случайные малые величины)
- количество иттераций (цикл по к=1,2,3...)
- берём случайный объект (один объект из нашей исходной выборки)
- только на этом объекте мы считаем градиент и только на этом объекте у нас считаются и изменяются веса. Улучшаются веса, значит улучшается предсказание. 
- при условии что веса перестали сильно изменяться можем останавливаться

Это классическое понимание стохастического градиентного спуска - один объект за раз. 

Есть разные модификации. Обучаемся не на одном объекта, а на десятке, на какой то доле объектов. 

стахастика в итоге должгно подойти к тому же показателю ошибки примерно как и градиентный спуск. Для стахастического метода требется большее количество иттераций. Но одна иттерация требует гораздо меньше времени и вычислений. 

Стахостический градиентный спуск хорошо подходит для больших данных так как не нужно весь массив данных загружать в оперативную память. 

## SGD своими руками<a class="anchor" id="sgd_manual"></a>

Реализуем стохастический градиентный спуск своими руками.

````python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
````

Берём данные для задачи регрессии, у нас все ещё регрессия. 1000 объектов возьмём (n_samples=1000), 2 признака (n_features=2), 2 полезных признака (n_informative=2), 1 целевое значение (n_targets=1), добавим немножко шума (noise=5), плюсом будем возвращать коэфиценты исходные (coef=True) настоящие коэфиценты которые должны быть для идеального решения данной задачи, фиксируем рандом стейт для воспроизводимости (random_state=2). 

````python
# сгенерируем набор данных
X, Y, coef = datasets.make_regression(n_samples=1000, n_features=2, n_informative=2, n_targets=1, 
                                      noise=5, coef=True, random_state=2)
X[:, 0] *= 10
display(X, Y, coef)
````

На выходе получаем Х - 2 признака и 1000 объектов. Один признак мы подшаманили и домножили на 10. `X[:, 0] *= 10` для дальнейшего шага масштабирования данных. У - целевые значения 1000 шт. и наши 2 веса в конце coef = `array([62.6735955, 68.0472182])`. 2 признака - 2 веса. 

Визуализируем наши данные. Эту выборку нам надо научиться предсказывать с помощью линейной регрессии. 

````python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], Y)

ax.set_xlabel('X0')
ax.set_ylabel('X1')
ax.set_zlabel('Y')
plt.show()
````

Отмасштабируем получившиеся признаки методом стандартизации.

Первым делом мы увидели, что признаки имеют разные масштабы, значит нужно их масштабировать, чтобы шрадиентный спуск обучался нормально. Сделаем стандартизацию, так как она себя показала лучше для линейных подходов. Линейная регрессия - линейный подход. 

Находим среднее по признакам `means = np.mean(X, axis=0)` у нас получается 2 объекта, 2 признака `array([-0.79012702,  0.02045785])`. 

Также находим стандартное отклонение `stds = np.std(X, axis=0)`для нашей выборки по оси 0. Получаем 2 значения стандартного отклонения.  

При этом на втором признаке среднее около 0 (0.02045785), а стандартное отклонение около 1 (0.99101454). То есть наш второй признак итак похож на стандартизированный. Можем добиться совсем идеала, чтобы среднее был 0, а стандартное отклонение было 1. 

Первый признак далеко не близок к стандартизированному, поэтому их необходимо пригнать к одинаковому масштабу. 

````python
# Получим средние значения и стандартное отклонение по столбцам

means = np.mean(X, axis=0)
stds = np.std(X, axis=0)
# параметр axis указывается для вычисления значений по столбцам, а не по всему массиву
display(means, stds)
````

````python
X.shape[0]
````

Для приведения к одинаковому масштабу, мы должны ппойтись по каждому нашему объекту, по каждому признаку и применить формулу `X[i][j] = (X[i][j] - means[j]) / stds[j]`. 

делаем проход по объектам `X.shape[0]` делаем проход по признакам `X.shape[1]` и стандартизируем одно значение за раз. 

````python
# вычтем каждое значение признака из среднего и поделим на стандартное отклонение
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        X[i][j] = (X[i][j] - means[j]) / stds[j]
````

Получаем наши изменёные признаки, и они меньшего масштаба, особенно первый. 

````python
X
````

Проверим их средние стандартное отклонение. видим значения близкие к 0, а единичка в стандартных отклонениях идеальная. Мы добились чего хотели: среднее в 0, а отклонение в 1. 

````python
means = np.mean(X, axis=0)
stds = np.std(X, axis=0)

display(means, stds)
````

Далее делаем MSE по классике, получив сначала наши истинные значения `y = X.dot(w)` из них вычитаем предсказанные значения, возводим в квадрат и делим на число значений `(sum((y - y_pred)**2)) / len(y)`. 

````python
# реализуем функцию, определяющую среднеквадратичную ошибку
def mserror(X, w, y_pred):
    y = X.dot(w)
    return (sum((y - y_pred)**2)) / len(y)
````

Подготовка данных и средств проверки закончена. Далее реализуем сам стохастический градиентный спуск. Будем основываться на реализации градиентного спуска. 

Первый шаг - инициализация наших весов. У нас весов 2 штучки. сколько признаков, столько и весов, поэтому мы генерируем нолики в количестве 2 шт. `w = np.zeros(X.shape[1])`. 

все наши веса будем записывать в список, чтобы потом к ним обратиться `w_list = [w.copy()]` и их поизучать. 

Также будем записывать значения ошибок `errors = []`, чтобы затем строить графики. 

Вводим шаг градиентного спуска `eta = 0.01`. 

Наше максимальное количество иттераций `max_iter = 1e3`

Критерий остановки по весам `min_weight_dist = 1e-8`. Когда веса перестают сильно изменяться, мы выходим из цикла обучения.  

Чтобы зайти в цикл обучения разница весов `weight_dist = np.inf`. Эта разница должна быть очень большой. 

Счётчик итераций должен быть нулевым `iter_num = 0` чтобы мы зашли в цикл.

Идём по нашему циклу `while weight_dist > min_weight_dist and iter_num < max_iter:` пока у нас большая дистанция между весами `weight_dist > min_weight_dist` и пока у нас допустимое количество иттераций `iter_num < max_iter` мы находимся в цикле. 

генерируем одно значение, один объект из обучающей выборке `train_ind = np.random.randint(X.shape[0], size=1)` и только на нём будем обучаться. Мы генерируем его индекс. 

Предсказание считаем только на сгенерированном элементе `y_pred = np.dot(X[train_ind], w)`. Получаем предсказание. 

Смотрим насколько предсказание близко к истине только по этому объекту `y_pred - Y[train_ind]`. Берём для подсчёта градиента только этот объект `np.dot(X[train_ind].T`. и делим на количество, а это один объект `Y[train_ind].shape[0]`.  Домножаем на скорость обучения `eta * `. Из текущих весов вычитаем то что у нас получилось, тем самым получая новый вес. 
` new_w = w - eta * 2 / Y[train_ind].shape[0] * np.dot(X[train_ind].T, y_pred - Y[train_ind])`  Обновляем веса на линейной моделе.

Получив новые веса, сравниваем с предыдущими `weight_dist = np.linalg.norm(new_w - w, ord=2)`. Находим их разницу и считаем L2 норму и получаем дистанцию между весами `weight_dist`. 

Уже на новых весах считаем ошибку. `error = mserror(X, new_w, Y)`.

Записываем веса и ошибку. `w_list.append(new_w.copy())`, `errors.append(error)`

Добавим немножко логирования. Для каждой сотой иттерации. 
`if iter_num % 100 == 0:
        print(f'Iteration #{iter_num}: W_new = {new_w}, MSE = {round(error, 2)}')`
        
Инкриментируем количество иттераций `iter_num += 1`. 

````python
%%time
# инициализируем начальный вектор весов
w = np.zeros(X.shape[1])

# список векторов весов после каждой итерации
w_list = [w.copy()]

# список значений ошибок после каждой итерации
errors = []

# шаг градиентного спуска
eta = 0.01

# максимальное число итераций
max_iter = 1e3

# критерий сходимости (разница весов, при которой алгоритм останавливается)
min_weight_dist = 1e-8

# зададим начальную разницу весов большим числом
weight_dist = np.inf

# счетчик итераций
iter_num = 0

np.random.seed(1234)

# ход градиентного спуска
while weight_dist > min_weight_dist and iter_num < max_iter:
    
    # генерируем случайный индекс объекта выборки
    train_ind = np.random.randint(X.shape[0], size=1)
    
    y_pred = np.dot(X[train_ind], w)
    new_w = w - eta * 2 / Y[train_ind].shape[0] * np.dot(X[train_ind].T, y_pred - Y[train_ind])

    weight_dist = np.linalg.norm(new_w - w, ord=2)
 
    error = mserror(X, new_w, Y)
    
    w_list.append(new_w.copy())
    errors.append(error)
    
    if iter_num % 100 == 0:
        print(f'Iteration #{iter_num}: W_new = {new_w}, MSE = {round(error, 2)}')

    iter_num += 1
    w = new_w
    
w_list = np.array(w_list)

print(f'Iter {iter_num}: error - {error}, weights: {new_w}')
print(f'В случае использования стохастического градиентного спуска ошибка составляет {round(errors[-1], 4)}')
````

Для данного метода ошибка может увеличиваться на некоторых иттерациях, но по общей тенденции она снижается. 

Для стохастического градиентного спуска мы увеличили максимальное число итераций (max_iter) до 1000, что естественно, так как из-за специфики метода для достижения сходимости нужно большее количество шагов.

````python
# Визуализируем изменение весов (красной точкой обозначены истинные веса, сгенерированные вначале)
plt.figure(figsize=(13, 6))
plt.title('Stochastic gradient descent')
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')

plt.scatter(w_list[:, 0], w_list[:, 1])
plt.scatter(coef[0], coef[1], c='r')
plt.plot(w_list[:, 0], w_list[:, 1])

plt.show()
````

Изучим как оно всё обучилось. Выше график весов. И в конце видно мы наинаем крутиться возле красной точке идеальных весов. 

````python
# Визуализируем изменение функционала ошибки
plt.plot(range(len(errors)), errors)
plt.title('MSE')
plt.xlabel('Iteration number')
plt.ylabel('MSE')
````

Как и в случае градиентного спуска, вектор весов приближается к истинному. При этом падает и ошибка.

Добиться лучшей скорости сходимости в методе стохастического градиентного спуска можно варьируя величину шага или используя методы, подбирающие ее адаптивно.

Среди преимуществ SGD можно выделить гораздо более быстрое вычисление одного шага по сравнению с обычным градиентным спуском и отсутствие необходимости хранить всю выборку в памяти при работе метода, что в свою очередь позволяет работать с очень большими выборками, которые невозможно поместить в память.

## Переобучение <a class='anchor' id='overfit'>

У нас есть задача, где нужно установить некоторую функцию. 

Чтобы понять смысл переобучения и недообучения, начнем с примера. Допустим, у нас есть исходная известная зависимость 3-го порядка:

$$f(x) = 0.6 - 13.2x - 5.3 x^{2} - 4.17x^{3}.$$

Реализуем ее в виде python-функции и построим график.

````python
def f(x):
    return 0.6 - 13.2 * x - 5.3 * x ** 2 - 4.17 * x ** 3
````

Изобразим визуально функцию. Мы хотим обучить задачу регрессии на восстановление этой функции. 

````python
dots = np.linspace(-10, 10, 100)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5000, 5000)
plt.xlim(-10,10)

plt.plot(dots, f(dots), color='g');
````

Теперь сгенерируем датасет из десяти случайных точек, подчиняющихся этой зависимости, с добавлением шума, чтобы наши точки отклонялись от идеального закона и нанесем на график. В наших настоящих задачах всегда есть шум. 

````python
np.random.seed(16)
x_data = np.random.uniform(-10, 10, 10)
f_data = [f(i) for i in x_data] + np.random.uniform(-1000, 1000, 10)
````

````python
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5000, 5000)
plt.xlim(-10, 10)

plt.plot(dots, f(dots), color='g')
plt.scatter(x_data, f_data);
````

Теперь попробуем создать модель, способную восстановить исходную зависимость. Самым примитивным так называемый __константный алгоритм__, который везде выдаёт одно значение. То есть модель вида 

$$a(x) = w_{0}.$$

Зависимость такой модели от признаков будет иметь вид горизонтальной прямой, что, очевидно, не обобщает нашу зависимость в должном виде (см. график ниже). Будем выдавать среднее по нашим данным `dots.mean()` - и это недообучение, когда модель не понимает, что происходит в данных. 

````python
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5000, 5000)
plt.xlim(-10, 10)

plt.plot(dots, f(dots), color='g')
plt.scatter(x_data, f_data)
plt.plot(dots, [dots.mean()]*len(dots), color='r');
````

Далее усложним семейство алгоритмов (усложняем модель), применив линейную регрессию, которая в случае одного признака будет иметь вид. Берём линейную модель, которая будет иметь наклон. Для этого появляется новый вес, который домножается на признак w1*x.

$$a(x) = w_{0} + w_{1}x.$$

Обучим соответствующую модель, применив для этого методы python "из коробки" для работы с линейной регрессией. Возьмём готовую линейную модель из библиотеки sklearn. 

Обучаемся на наших данных 

````python
from sklearn.linear_model import LinearRegression

# создадим модель
linear_regressor = LinearRegression()

# обучим ее
linear_regressor.fit(np.reshape(x_data, (-1, 1)), f_data)
````

Посмотрем на наши веса - коэфицент наклона `linear_regressor.coef_[0]` -237.82 и на сдвиг `linear_regressor.intercept_` -411.92. 

````python
# выведем полученный вес при признаке и свободный коэффициент
print(linear_regressor.coef_[0], linear_regressor.intercept_)
````

Нанесем полученную после обучения модель на график

````python
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5000, 5000)
plt.xlim(-10, 10)

plt.plot(dots, f(dots), color='g')
plt.scatter(x_data, f_data)
plt.plot(dots, linear_regressor.predict(np.reshape(dots, (-1, 1))), color='r');
````

Лучше чем константная модель, появился наклон и сдвиг. Но наша модель простая - линейная и ожидать от неё, что она подстроится под изгибы нашей функции не стоит. она к этому не преспособлена. Метрики будут лучше чем у константной модели. 

Мы обучили линейную модель, и видим, что она, как и константная, плохо восстанавливает исходную зависимость. В данном случае можно говорить о __недообучении__. Хороший алгоритм не был построен, поскольку с помощью выбранного семейства алгоритмов невозможно восстановить исходную закономерность.

Усложним используемое семейство алгоритмов до кубической зависимости


## Методы борьбы с переобучением <a class='anchor' id='methods'>

Видно, что новая модель лучше описывает имеющиеся в обучающей выборке данные и дает фактически идеальные ответы на них, но про этом в целом зависимость сильно отличается от истинной. Поэтому если мы попробуем применить эту модель на новых данных, ответы будут расходиться с правильными. Такое явление и называется __переобучением__. Алгоритм слишком сильно подогнан под обучающую выборку, и за счет этого будет давать неадекватные ответы на новых точках.

Таким образом, недообучение несет за собой плохое качество на обучении и на новых данных, а переобучение - хорошее качество на обучении и плохое на новых данных.

Понятно, как бороться с недообучением - усложнять семейство алгоритмов. Возникает вопрос, как выявить переобучение и его избежать. В случае переобучения, как было сказано ранее, данные из обучающей выборки алгоритмом будут описываться хорошо, а новые данные - плохо, поэтому используя только обучающую выборку, невозможно заключить, хорошо обучен алгоритм или переобучен, так как оба они будут хорошо описывать известные данные.

Есть несколько методов оценки качества алгоритма и выявления переобучения:

1. Не использовать всю выборку для обучения, а откладывать часть данных для проверки полученного алгоритма. Это называется **отложенной выборкой**. Данные делятся на обучающую и тестовую выборку в соотношении, например, 0.7 к 0.3, и затем на первой части алгоритм обучается, а на второй проверяется. Размер отложенной выборки в данном случае нужно подбирать с осторожностью, так как слишком маленькая тестовая выборка не будет обеспечивать должной точности оценки качества обучения, а слишком маленькия обучающая выборка приведет к снижению качества обучения, так как будет малорепрезентативна. Таким образом, главный минус этого метода - сильная зависимость результата от того, как мы выбираем отложенную выборку. Например, в пространстве объектов могут быть какие-то особые, отличающиеся от остальных по какому-то свойству объекты, и может так произойти, что после разбиения они не попадут в обучающую выборку, алгоритм на них не обучится, и качество обучения после проверки на этих объектах, попавших в тестовую выборку, будет плохим. Один из путей решения этой проблемы - многократное случайное разбиение выборки на тестовую и обучающую и использование в качестве оценки качества среднего значения ошибки полученной после каждого разбиения. Но и этот метод не гарантирует, что каждый объект побывает в обучающей выборке, так как разбиения случайные.
<img src="images/valid.png" style="width: 500px;">


2. **Кросс-валидация** (усложненная версия метода отложенной выборки). Этот метод как раз вытекает из проблемы, описанной выше. Он является более системным подходом. В этом случае выборка разбивается на $k$ блоков, и затем каждый из них по очереди используется в качестве тестового, а остальная часть - в качестве обучающей выборки. После прохождения всей выборки таким образом получается $k$ показателей качества, и итоговая оценка качества обучения по кросс-валидации оценивается как средняя из этих $k$. В этом случае мы гарантируем, что все данные поучаствуют в обучении. Выбор количества блоков $k$ обычно зависит от размера выборки. Чем больше данных, тем меньше нужно блоков, так как во-первых в этом случае после разбиения даже на малое количество блоков у нас остается большой объем данных в обучающей выборке, что обеспечивает хорошее качество обучения, а во-вторых, разбиение на $k$ блоков означает обучение алгоритма $k$ раз, соответственно, чем их больше, тем больше получается вычислительная сложность процесса обучения модели. Обычно $k$ принимает значение от 3 до 10.

### Регуляризация <a class='anchor' id='reg'>

#### $L_2$-регуляризация (ridge, регуляризация Тихонова)

Метод регуляризации заключается в "штрафовании" модели за слишком большие веса путем добавления нового члена к ошибке (MSE):

$$Q(w, X) + \lambda ||w||^{2} \rightarrow \underset{w}{\text{min}}.$$

добавленный член $\lambda ||w||^{2}$ - **квадратичный регуляризатор**, который представляет собой $L_{2}$-норму вектора весов, то есть сумму квадратов весов $\sum^{d}_{j=1}w_{j}^{2}$, коэффициент $\lambda$ при нем - коэффициент регуляризации. Чем больше его значение, тем меньшая сложность модели будет получаться в процессе такого обучения. Если _увеличивать_ его, в какой-то момент оптимальным для модели окажется зануление всех весов. В то же время при слишком _низких_ его значениях появляется вероятность чрезмерного усложнения модели и переобучения. Выбор оптимального значения этого коэфициента является отдельной задачей и заключается в многократном обучении модели с разными его значениями и сравнении их качества.

Берём наши веса, возводим каждый вес в квадрат, суммируем по нашей L2 норме и берём корень квадратный. квадрат сокращает корень и остаеётся формула. 

По сути, смысл регуляризации заключается, как и в обычном обучении, в минимизации функционала ошибки, только в данном случае добавляется условие непревышения нормой вектора весов некоторого значения $||w||^{2}\leq C$, то есть ограничение весов, что и будет залогом избежания переобучения.

$$\begin{cases} Q(w, X) \rightarrow min \\ ||w||^2 \leq C \end{cases}$$

$$C = \frac{1}{\lambda}$$

````python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])

y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
````

Возьмём данные с предыдущего урока, где надо было предсказать баллы по егэ в зависимости от стажа преподавателя. В первом столбце наш вспомогательный признак для сдвига, для интерсепта для свободного коэфицента.  Целевые значения теже самые для длинейной регрессии. 
Посмотрим что будет происходить с нашей моделью, с её весами в зависимости от коэфицента регуляризации. alpha - это лямда из формул выше. 


На графике ниже изображено изменение весов признаков при увеличении коэффициента регуляризации `alpha` от $10^{-3}$ до $100$ для модели `Ridge` (L2-регуляризация). Чем больше значение `alpha`, тем сильнее регуляризация и тем сильнее модель "штрафует" за большие значения весов, они убывают плавно и примерно одновременно друг с другом.


````python
np.logspace(-3, 2, 50)
````

Нарисуем графики. Что будет происходить с нашими весами в зависимости от коэфицента регуляризации. Возьмём Ridge регуляризацию, это наша линейная модель, но там встроена регуляризация L2. 

Разбиваем данные на 2 куска. По 5 объектов 

```
train_amount = 5
train_X = X[:train_amount]
train_y = y[:train_amount]
```
Будем наши коэфиценты записывать, их у нас 50 `alpha_list = np.logspace(-3, 2, n)`. У нас 2 признака, сколько признаков столько будет и весов `train_X.shape[1]`. Один вес для сдвига, один вес настоящий (для настоящего признака). Эксперементируем с такими весами `np.logspace(-3, 2, 50)` нагенерировали 50 разных коэфицентов регуляризации.

будем по ним ходить в цикле и обучать Ridge регуляризацию. При этом будем постоянно менять коэфицент регуляризации. `Ridge(alpha=val)` взяли первый, обучили модель, взяли второй обучили модель... 

`fit_intercept=False` запрещаем модели создавать дополнительный коэфицент для сдвига. У нас уже есть свой сдвиг (первая колонка в данных). Явна просим модель не делать дополнительный вес. По умолчанию этот параметр True. 

`ridge =` - получили модель. Дальше модель обучается на первом коэфиценте `alpha=val`, на первой иттерации `ridge.fit(train_X, train_y)`. Самый маленький коэфицент регуляризации говорит о том, что у модели большая свобода действий - она может брать очень большие веса. 

` coeffs[i, :] = ridge.coef_.flatten()` коэфиценты будем добавлять в список.  

```
for i in range(train_X.shape[1]):
    plt.plot(alpha_list, coeffs[:, i])
```

Будем отрисовывать что у нас получается. 
Видим что при очень маленьких коэфицентах регуляризации - веса большие. и с увеличением коэфицента (оранжевая линия), веса уменьшаются (синяя линия).

````python
from sklearn.linear_model import Ridge

n = 50

train_amount = 5
train_X = X[:train_amount]
train_y = y[:train_amount]
test_X = X[train_amount:]
test_y = y[train_amount:]


coeffs = np.zeros((n, train_X.shape[1]))
alpha_list = np.logspace(-3, 2, n)

for i, val in enumerate(alpha_list):
    ridge = Ridge(alpha=val, fit_intercept=False)
    ridge.fit(train_X, train_y)
    
    coeffs[i, :] = ridge.coef_.flatten()

for i in range(train_X.shape[1]):
    plt.plot(alpha_list, coeffs[:, i])

plt.title('Убывание абсолютных значений весов признаков\n при увеличении коэффициента регуляризации alpha (Ridge)')
plt.xticks(np.arange(0, 101, 10))
plt.xlabel('alpha')
plt.ylabel('Вес признака');
````

Поискать самый лучший коэфицент регуляризации. Для этого возьмём наш список коэфицентов `coef = coeffs[index]`.  V s все коэфиценты добавляли в заранее определённый массив `coeffs[i, :] = ridge.coef_.flatten()`.  Будем доставать эти коэфиценты и считать предсказания.  Начнём с нуля index = 0.

Меняем коэфицент регуляризации и смотрим на показатели index = 10
Снова меняем коэфицент и смотрим снова. таким образом ищем оптимальные сочитания, чтобы метрики на тесте и на обучении были приемлимые. 

````python
index = 0

coef = coeffs[index]

print(f'Коэффициент регуляризации: {alpha_list[index]}')
print(f'Веса: {coef}')

fig, axs = plt.subplots(1, 2, figsize=(15, 4))
ax1, ax2 = axs

ax1.scatter(train_X[:, 1], train_y)

y_pred = np.dot(train_X, coef)
ax1.plot(train_X[:, 1], y_pred, c='r')

ax1.set_xlabel('X1')
ax1.set_ylabel('Y')
print(f'MSE на обучении: {np.mean((y_pred - train_y)**2)}')

ax2.scatter(test_X[:, 1], test_y)

y_pred = np.dot(test_X, coef)
ax2.plot(test_X[:, 1], y_pred, c='r')

ax2.set_xlabel('X1')
ax2.set_ylabel('Y')
print(f'MSE на тесте: {np.mean((y_pred - test_y)**2)}')
plt.show()
````

#### $L_1$-регуляризация (lasso, регуляризация через манхэттенское расстояние)

Описанный выше метод с использованием $L_{2}$-нормы вектора весов в качестве регуляризатора называется **$L_{2}$-регуляризацией**. По аналогии существует также **$L_{1}$-регуляризация**, использующая в качестве регуляризатора $L_{1}$-норму вектора весов, то есть сумму модулей весов.

$$||w||_{1} = \sum^{d}_{j=1}|w_{j}|.$$

На графике ниже изображено изменение весов признаков при увеличении коэффициента регуляризации `alpha` от $10^{-3}$ до $100$ для модели `Lasso` (L1-регуляризация). Чем больше значение `alpha`, тем сильнее регуляризация и тем сильнее модель "штрафует" за большую абсолютную величину признаков. 

Такой метод часто используется для отбора признаков: у менее ценных признаков гораздо раньше обнуляются веса.

### с лекции
Суть - контролировать модель обучения. но используется другой регуляризатор. Можно назвать его абсолютным, так как используется модуль весов. Берём наши еса, берём  уних модуль и суммируем наши веса. 

Побробуем также используя разные коэфиценты регуляризации и пообучаем модель лассо. 
будем записывать все коэфиценты `coeffs = np.zeros((n, train_X.shape[1]))`. Будем ходить по нашим коэфицентам регуляризации `alpha_list`, обучать лассо на разных коэфицентах регуляризации ` lasso = Lasso(alpha=val, fit_intercept=False)`.
Обучаемся `lasso.fit(train_X, train_y)`. Добавляем коэфиценты `coeffs[i, :] = lasso.coef_.flatten()`. И отрисовываем результаты. 



````python
from sklearn.linear_model import Lasso

n = 50

coeffs = np.zeros((n, train_X.shape[1]))
alpha_list = np.logspace(-3, 2, n)

for i, val in enumerate(alpha_list):
    lasso = Lasso(alpha=val, fit_intercept=False)
    lasso.fit(train_X, train_y)
    coeffs[i, :] = lasso.coef_.flatten()

for i in range(train_X.shape[1]):
    plt.plot(alpha_list, coeffs[:, i])

    
plt.title('Убывание абсолютных значений весов признаков\n при увеличении коэффициента регуляризации alpha (Lasso)')
plt.xticks(np.arange(0, 101, 10))
plt.xlabel('alpha')
plt.ylabel('Вес признака');
````

Также можем походить по разным коэфицентам регуляризации. 

````python
index = 35

coef = coeffs[index]

print(f'Коэффициент регуляризации: {alpha_list[index]}')
print(f'Веса: {coef}')

fig, axs = plt.subplots(1, 2, figsize=(15, 4))
ax1, ax2 = axs

ax1.scatter(train_X[:, 1], train_y)

y_pred = np.dot(train_X, coef)
ax1.plot(train_X[:, 1], y_pred, c='r')

ax1.set_xlabel('X1')
ax1.set_ylabel('Y')
print(f'MSE на обучении: {np.mean((y_pred - train_y)**2)}')

ax2.scatter(test_X[:, 1], test_y)

y_pred = np.dot(test_X, coef)
ax2.plot(test_X[:, 1], y_pred, c='r')

ax2.set_xlabel('X1')
ax2.set_ylabel('Y')
print(f'MSE на тесте: {np.mean((y_pred - test_y)**2)}')
plt.show()
````

### Коэффициент детерминации

_Коэффициент детерминации_ $R^{2}$ является еще одной метрикой качества в задачах регрессии. Ранее мы говорили о средней абсолютной и среднеквадратичной ошибке. Коэффициент детерминации позволяет развить тему среднеквадратичной ошибки, интерпретируя ее. 

MSE не позволяет сама по себе сделать вывод о том, как хорошо модель решает задачу. Например, если целевая переманная принимает значения от 0 до 1, а MSE равняется 10, это плохой показатель, а когда целевая переменная варьируется от 1000 до 10000, такое же значение уже является очень хорошим. Для избавления от такой неясности и был введен коэффициент детерминации, который по сути является нормированной среднеквадратичной ошибкой и принимает значения от 0 до 1.

$$R^2 = 1 - \frac{\frac{1}{n}\sum^{l}_{i=1}{(y - y_{pred})^2}}{\frac{1}{n}\sum^{l}_{i=1}{(y - \bar{y})^2}}$$

где $\bar{y}=\frac{1}{l}\sum^{l}_{i=1}y_{i}$ - среднее значение целевой переменной.

Коэффициент детерминации характеризует, какую долю дисперсии ответов объясняет модель. Если $R^{2}=1$, то модель идеально описывает данные, если же $R^{2}$ близко к нулю, то предсказания сопоставимы по качеству с константной моделью.

Вспомнив MSE. Это отклонение бывает разным. Как 2 примера снизу. Бывает маленькое отклонение 0.04, бываетбольшое 400. В первом примере целевое значение маленькое и для маленькой задачи отклонение 0.04 может быть большим. А если для маленькой задачи отклонение будет 400 - это будет очень много. Гигантское значение ошибки. При этом взяв целевое значение побольше, то отклонение в 0.04 это очень-очень маленькое отклонение, мы очень здорово решили задачу. А отклонение 400 - среднее отклонение. 

Вроде бы метрика универсальная, но в разных задачах, с разными целевыми значениями разное MSE даёт нам разные выводы. Могут возникать некоторые смущения. Нет границы плохого показателя. 

Тут мы подходим к метрике, которая везде одинакова - метрика R2 - некая точность для задач регрессии. Из единички вычитаем дробь, где в знаменателе у нас получается дисперсия, а в числителе MSE. 

Для двух задач снизу коэфицент детерминации будет сопостовим. 

````python
true = np.array([0, 0.2, 0.1, 0.6, 0.3, 0.9, 0.7])
pred = np.array([0.1, 0.3, 0.2, 0.7, 0.5, 0.5, 0.5])

np.mean((true - pred) ** 2)
````

Нормализация, где мы работаем с минимумом и максимумом. Берём минимум и максимум и скалируем, масштабируем. С каждого признака вычитаем минимум и делим на разницу  максимум - минимум.

На всякий случай делаем копию датасета и проводим нормализацию. Так как данные у нас int, чтобы признаки лучше считались приводим их к float. Нампай может не перевести сам во флоат, поэтому делаем это принудительно. 

````python
X_norm = X.copy()
X_norm = X_norm.astype(np.float64)
X_norm
````

Масштабируем все наши признаки

````python
X_norm[:, 1] = min_max_scale(X_norm[:, 1])
X_norm[:, 2] = min_max_scale(X_norm[:, 2])
X_norm[:, 3] = min_max_scale(X_norm[:, 3])
X_norm
````

Стандартизация.
Меняем среднее и стандартное отклонения. 

В первом признаке среднее у нас 3.0

В стандартизациии мы должны посчитать среднее и стандартное отклонение. Из признака вычитаем среднее и делим на стандартное отклонение. 

````python
def standard_scale(X):
    mean = X.mean()
    std = X.std()
    return (X - mean) / std
````

переводим во float и делаем стандартизацию наших признаков. Теперь 0 - это у нас среднее, а 1 это разбег, разброс. 

````python
X_st = X.copy().astype(np.float64)
X_st[:, 1] = standard_scale(X_st[:, 1])
X_st[:, 2] = standard_scale(X_st[:, 2])
X_st[:, 3] = standard_scale(X_st[:, 3])

X_st
````

### SGD (Stochastic gradient descent)

функция для ошибки MSE

````python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
````

Инициализируем веса. Берём случайные маленькие величины.
4 признака = 4 веса. Первый признак для отклонения - сдвиг. Для фиктивных единичек. 

````python
W = np.random.randn(X.shape[1])
W
````

Начнём с классического градиентного спуска. Каждый раз при создании функции веса будут инициализироваться заново `W = np.random.randn(X.shape[1])`. 

Берём количество наших объектов `n = X.shape[0]`

Проходимся по желаемому количеству иттераций `for i in range(0, iterations):`. 

Получаем предсказания `y_pred = np.dot(X, W)`.

Считаем ошибку от этих предсказаний `err = calc_mse(y, y_pred)`.

Делаем градиент `dQ = 2/n * X.T @ (y_pred - y)`.

Умножаем градиент на скорость обучения и вычитаем из текущех весов это произведение, тем самым получаем новые веса `W -= (eta * dQ)`.

`print(f'Iter: {i}, weights: {W}, error {err}')` - добавляем логгирование для каждой 10 иттерации.

`calc_mse(y, np.dot(X, W))` - считаем финальную ошибку.

Возвращаем веса. `return W`

````python
# классический  GD
def gradient_descent(X, y, iterations, eta=1e-4):
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    for i in range(0, iterations):
        y_pred = np.dot(X, W)
        err = calc_mse(y, y_pred)
        dQ = 2/n * X.T @ (y_pred - y) # градиент функции ошибки
        W -= (eta * dQ)
        if i % (iterations / 10) == 0:
            print(f'Iter: {i}, weights: {W}, error {err}')
    print(f'Final MSE: {calc_mse(y, np.dot(X, W))}')
    return W
````

Пообучаемся на масштабированных признаках. 5000 иттераций и скорость обучения 0.01

````python
gradient_descent(X_st, y, iterations=5000, eta=1e-2)
````

Модифицируем наш обычный градиентный спуск и сделаем стохастический. 

Берём наши веса и количество объектов `W = np.random.randn(X.shape[1])`,`n = X.shape[0]`

Проходимся по иттерациям `for i in range(0, iterations):`

И вносим изменения. берём один объект в том количестве, которое мы себе указали. `inds = np.random.randint(n, size=batch_size)`. 

берём только эти признаки и только эти целевые значения 
```
X_tmp = X[inds, ]
y_tmp = np.array(y)[inds]
```

затем мы считаем предсказания на объекте `y_pred_tmp = np.dot(X_tmp, W)`. 

считаем наш градиент только на этом объекте `dQ = 2/len(y_tmp) * X_tmp.T @ (y_pred_tmp - y_tmp)`

получаем нвые веса `W -= (eta * dQ)`

считаем ошибку `err = calc_mse(y, np.dot(X, W))`

````python
# стохастический градиентный спуск
def stohastic_gradient_descent(X, y, iterations, batch_size, eta=1e-4):
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    n_batch = n // batch_size    
    if n % batch_size != 0:
        n_batch += 1
    print(f'amount of batches is {n_batch}')
        
    for i in range(0, iterations):
        
        for b in range(n_batch):
            inds = np.random.randint(n, size=batch_size)

            X_tmp = X[inds, ]
            y_tmp = np.array(y)[inds]

            
            y_pred_tmp = np.dot(X_tmp, W)
            dQ = 2/len(y_tmp) * X_tmp.T @ (y_pred_tmp - y_tmp) # градиент функции ошибки
            W -= (eta * dQ)
            
            err = calc_mse(y, np.dot(X, W))
        
        if i % (iterations / 10) == 0:
            print(f'Iter: {i}, weights: {W}, error {err}')
    
    print(f'Final MSE: {calc_mse(y, np.dot(X, W))}')
    return W
````

````python
stohastic_gradient_descent(X_st, y, iterations=5000, batch_size=4, eta=1e-2)
````

##### L1 регуляризация

$$Q(w, X) + \lambda |w| \rightarrow \underset{w}{\text{min}}.$$
Про производную можно посмотреть [здесь](https://youtu.be/dHhYHGI9E6I).

##### L2 регуляризация

$$Q(w, X) + \lambda ||w||^{2} \rightarrow \underset{w}{\text{min}}.$$

<center>Производная</center>

$$d\lambda ||w||^{2} = 2\lambda w $$

L2 реуляризация добавляется к ошибке. Нам надо будет найти производную по квадратичному регуляризатору.

Модифицируем обычный градиентный спуск с L2 регуляризацией. 

На вход принимаем регуляризацию для нашего регуляризатора. `reg=1e-8`.  
Получаем веса `W = np.random.randn(X.shape[1])`, проходимся по иттерациям `for i in range(0, iterations):`. 
Предсказания, ошибка, градиент по ошибке, градиент по регуляризации. Чобы не писать двойку, так как это константа можем подумать что она зашита в коэфицент регуляризации `reg`. 

Наш суммарный градиент - это градиент по ошибке + градиент по регуляризатору. `(dQ + dReg)`. Эту сумму умножаем на скорость обучения и вычитаем из весов, получая новые веса. По суммарной ошибке изменяем веса. 

````python
def gradient_descent_reg_l2(X, y, iterations, eta=1e-4, reg=1e-8):
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    for i in range(0, iterations):
        y_pred = np.dot(X, W)
        err = calc_mse(y, y_pred)
        
        dQ = 2/n * X.T @ (y_pred - y) # градиент функции ошибки
        dReg = reg * W # градиент регуляризации
        
        W -= eta * (dQ + dReg)
        
        if i % (iterations / 10) == 0:
            print(f'Iter: {i}, weights: {W}, error {err}')
    
    print(f'Final MSE: {calc_mse(y, np.dot(X, W))}')
    return W
````

````python
gradient_descent_reg_l2(X_st, y, iterations=5000, eta=1e-1, reg=1e-4)
````

````python
gradient_descent(X_st, y, iterations=5000, eta=1e-1)
````

## Summary

_Масштабирование_
* Масштабирование признаков - хорошая практика, позволяющая обучать модели быстрее и делающая их более точными
* При использовании метрических алгоритмов масштабирование обязательно (!)

_Стохастический градиентный спуск_
* Стохастический градиентный спуск (SGD) - на каждом шаге уменьшаем ошибку только на одном объекте (или нескольких), а не на всей выборке, работает быстрее обычного GD, т.к. меньше вычислений
* SGD из-за стохастичности может "перепрыгнуть" локальный минимум и попасть в глобальный

_Регуляризация_
* Переобучение - модель "выучила" выборку, но обобщающая способность слабая
* Признаки переобучения: качество на трейне высокое, а на тесте низкое, большие веса модели
* Один из способов борьбы с переобучением - регуляризация - штраф за большие веса

### Определения
*Масштабирование данных*

**Нормализация данных** — метод предобработки числовых признаков в обучающих наборах данных с целью приведения их к некоторой общей шкале без потери информации о различии диапазонов. (шкала от 0 до 1)

**Стандартизация** — метод предобработки с целью приведения данных к единому формату и представлению. (М=0, std=1)
___________
_Стохастический градиентный спуск_

**Стохастический градиентный спуск** — метод нахождения локального экстремума функции (минимума или максимума) с помощью движения вдоль градиента, который считается на каждом шаге не как сумма градиентов от каждого элемента выборки, а как градиент от одного, случайно выбранного элемента.

___________
_Переобучение_

**Переобучение** (overfitting) — явление, когда алгоритм хорошо объясняет примеры из обучающей выборки (обеспечивает малую величину ошибки), но плохо работает на примерах, не участвовавших в обучении (не обеспечивает малую величину ошибки).

**Недообучение** (underfitting) — явление, при котором алгоритм обучения не обеспечивает достаточно малой величины средней ошибки на обучающей выборке.

**Кросс-валидация** — процедура оценивания обобщающей способности алгоритмов. С её помощью эмулируется наличие тестовой выборки, которая не участвует в обучении, но для которой известны правильные ответы.

________
_Регуляризация_

**Регуляризация** — метод добавления некоторых дополнительных ограничений к условию с целью решить некорректно поставленную задачу или предотвратить переобучение. Эта информация часто имеет вид штрафа за сложность модели.
$$Q(w, X) + \lambda ||w||^{2} \rightarrow \underset{w}{\text{min}}.$$

**L1-регуляризация** (lasso, регуляризация через манхэттенское расстояние) — метод добавления дополнительных ограничений в виде $L_{1}$-норму вектора весов, то есть сумму модулей весов. 
$$||w||_{1} = \sum^{d}_{j=1}|w_{j}|.$$

**L2-регуляризация** (ridge, регуляризация Тихонова) — метод добавления дополнительных ограничений в виде $L_{2}$-норму вектора весов, то есть сумму квадратов весов. 
$$||w||_{2} = \sum^{d}_{j=1}w^{2}.$$
